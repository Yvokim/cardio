```python
#The difference in total cases between Italy and Germany exceeded 100,000

g = df[df['Country']=='Germany'].sort_values(by='Date').reset_index(drop=True)
i=df[df['Country']=='Italy'].sort_values(by='Date').reset_index(drop=True)

min_length=min(len(g),len(i))
diff=abs(g.loc[:min_length-1,'Total Cases'] -i.loc[:min_length-1, 'Total Cases'])


if any(diff>100000): 
    date_exceeds_100k= g.loc[diff>100000, 'Date'].iloc[0] 
    
else :
 'Not Found'

print(date_exceeds_100k)



```

```python
# What country has the highest 3rd death rate

country_death_rate=df.groupby('Country')['Deaths per Million'].mean().reset_index()

sorted= country_death_rate.sort_values(by='Deaths per Million',ascending=False)

sorted = sorted.reset_index(drop=True)
third_highest_country=sorted.loc[2,'Country']
third_highest_rate=sorted.loc[2,'Deaths per Million']

print(third_highest_country)
print(third_highest_rate)




```

```python
# What is the F1 score of the following statement,, Countries where more than 17% the population is over 65 years old,have death rates over 2000 per million inhabitants.
# True positive: fulfills both conditons
#True negative: Fulfills neither condition
#False positive: Fulfills condition one but not 2
#False neg: Fulfills 2 but not one

byCountries= df.groupby('Country')[['Deaths per Million','Population Over 65 (%)']].mean().reset_index()

print(byCountries)
TP=byCountries[(byCountries['Population Over 65 (%)']>17)  & (byCountries['Deaths per Million']>2000)]
print(f'True positive:\n {TP}')
TPCount= TP.shape[0]


TN=byCountries[(byCountries['Population Over 65 (%)']<17) & (byCountries['Deaths per Million']<2000)]
print(f'True Negative:\n{TN}')
TNCount = TN.shape[0]

FP= byCountries[(byCountries['Population Over 65 (%)']>17) &(byCountries['Deaths per Million']<2000)]
print(f'False Positive:\n{FP}')
FPCount= FP.shape[0]

FN= byCountries[(byCountries['Population Over 65 (%)']<17) &(byCountries['Deaths per Million']>2000)]
print(f'False Negative:\n{FN}')
FNCount= FN.shape[0]

precision= (TPCount/(TPCount + FPCount))
Recall = (TPCount/(TPCount + FNCount))

F1= 2*(precision*Recall)/(precision+Recall)

print(f'F1 score:\n{F1}')
```
