```python
std_h=df['height'].std()
upper=df.height.mean()+(2*std_h)
lower=df.height.mean()-(2*std_h)

outliers=df[(df.height<lower)|(df.height>upper)]

num_outliers=outliers.shape[0]
perc=(num_outliers/df.shape[0])*100
print(f'{perc} %')
```
